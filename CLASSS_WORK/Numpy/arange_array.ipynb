```python
import numpy as np

# Create an array with values from 0 to 9
arranged_numbers = np.arange(10)
print("Array from 0 to 9:", arranged_numbers)

# Create an array with values from 5 to 14
arranged_numbers_start_end = np.arange(5, 15)
print("Array from 5 to 14:", arranged_numbers_start_end)

# Create an array with values from 0 to 9 with a step of 2
arranged_numbers_step = np.arange(0, 10, 2)
print("Array from 0 to 9 with step 2:", arranged_numbers_step)
```
