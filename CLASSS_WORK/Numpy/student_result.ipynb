```python
# Smart Student Result Analyzer (Without CSV Module)

def assign_grade(percentage):
    if percentage >= 90:
        return "A+"
    elif percentage >= 80:
        return "A"
    elif percentage >= 70:
        return "B"
    elif percentage >= 60:
        return "C"
    elif percentage >= 50:
        return "D"
    else:
        return "F"


def analyze_results():
    toppers = {}

    try:
        file = open("students.txt", "r")

        lines = file.readlines()

        # Skip header line
        for line in lines[1:]:
            try:
                data = line.strip().split(",")

                # Check missing data
                if len(data) < 6:
                    raise ValueError("Missing fields")

                student_id = data[0]
                name = data[1]
                student_class = data[2]

                # Convert marks
                m1 = float(data[3])
                m2 = float(data[4])
                m3 = float(data[5])

                # Validate marks
                if m1<0 or m1>100 or m2<0 or m2>100 or m3<0 or m3>100:
                    raise ValueError("Invalid marks")

                # Calculate result
                total = m1 + m2 + m3
                percentage = total / 3
                grade = assign_grade(percentage)

                print("ID:", student_id)
                print("Name:", name)
                print("Class:", student_class)
                print("Total:", total)
                print("Percentage:", round(percentage,2), "%")
                print("Grade:", grade)
                print("-------------------")

                # Find topper per class
                if student_class not in toppers:
                    toppers[student_class] = (name, percentage)

                elif percentage > toppers[student_class][1]:
                    toppers[student_class] = (name, percentage)

            except ValueError as e:
                print("Skipping record:", line.strip(), "-", e)

        file.close()

        # Display toppers
        print("\nClass Toppers:")
        for cls in toppers:
            print("Class", cls, "Topper:", toppers[cls][0],
                  "-", round(toppers[cls][1],2), "%")

    except FileNotFoundError:
        print("File not found!")


# Run Program
analyze_results()
```
