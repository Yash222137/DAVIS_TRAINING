```python
#Question 1: A shopkeeper wants to calculate total bill after discount.
#Input: Enter price and discount percentage
#Output: Final price after discount
price = float(input("Enter the original price: "))
discount_percentage = float(input("Enter the discount percentage: "))

discount_amount = (discount_percentage / 100) * price
final_price = price - discount_amount

print(f"The final price after a {discount_percentage}% discount is: ${final_price:.2f}")
```

```python
#Question 2: Check whether a person is eligible to vote.
#Input: Enter age
#Output: Eligible / Not Eligible

age = int(input("Enter the age: "))

if age >= 18:
    print("Eligible")
else:
    print("Not Eligible")
```

```python
#Question 3: Calculate electricity bill based on units consumed.
#Input: Enter units
#Output: Total bill amount

units = float(input("Enter the units consumed: "))

total_bill = 0.0

# Define slab rates (example rates)
if units <= 50:
    total_bill = units * 3.50
elif units <= 150:
    total_bill = 50 * 3.50 + (units - 50) * 4.00
elif units <= 250:
    total_bill = 50 * 3.50 + 100 * 4.00 + (units - 150) * 5.20
else:
    total_bill = 50 * 3.50 + 100 * 4.00 + 100 * 5.20 + (units - 250) * 6.50

print(f"The total electricity bill for {units} units is: ${total_bill:.2f}")
```
