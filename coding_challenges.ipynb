```python
#Question 1: A shopkeeper wants to calculate total bill after discount.
#Input: Enter price and discount percentage
#Output: Final price after discount
price = float(input("Enter the original price: "))
discount_percentage = float(input("Enter the discount percentage: "))

discount_amount = (discount_percentage / 100) * price
final_price = price - discount_amount

print(f"The final price after a {discount_percentage}% discount is: ${final_price:.2f}")
```

```python
#Question 2: Check whether a person is eligible to vote.
#Input: Enter age
#Output: Eligible / Not Eligible

age = int(input("Enter the age: "))

if age >= 18:
    print("Eligible")
else:
    print("Not Eligible")
```

```python
#Question 3: Calculate electricity bill based on units consumed.
#Input: Enter units
#Output: Total bill amount

units = float(input("Enter the units consumed: "))

total_bill = 0.0

# Define slab rates (example rates)
if units <= 50:
    total_bill = units * 3.50
elif units <= 150:
    total_bill = 50 * 3.50 + (units - 50) * 4.00
elif units <= 250:
    total_bill = 50 * 3.50 + 100 * 4.00 + (units - 150) * 5.20
else:
    total_bill = 50 * 3.50 + 100 * 4.00 + 100 * 5.20 + (units - 250) * 6.50

print(f"The total electricity bill for {units} units is: ${total_bill:.2f}")
```

```python
#Question 5: Create a function to calculate simple interest.
#Input: Enter P, R, T
#Output: Simple Interest

def calculate_simple_interest(principal, rate, time):
    """
    Calculates the simple interest.

    Args:
        principal (float): The principal amount.
        rate (float): The annual interest rate (as a percentage).
        time (float): The time in years.

    Returns:
        float: The simple interest.
    """
    # Simple Interest = (P * R * T) / 100
    simple_interest = (principal * rate * time) / 100
    return simple_interest

# Get input from the user
P = float(input("Enter the principal amount (P): "))
R = float(input("Enter the annual interest rate (R in %): "))
T = float(input("Enter the time in years (T): "))

# Calculate and print the simple interest
interest = calculate_simple_interest(P, R, T)
print(f"The simple interest is: {interest:.2f}")
```

```python
#Question 4: Print all even numbers between 1 to N.
#Input: Enter N
#Output: Even numbers list

N = int(input("Enter a number (N): "))

print(f"Even numbers between 1 and {N}:")
for i in range(1, N + 1):
    if i % 2 == 0:
        print(i)
```

```python
#Question 6: A shopkeeper wants to calculate total bill after discount.
#Input: Enter price and discount percentage
#Output: Final price after discount

price = float(input("Enter the original price: "))
discount_percentage = float(input("Enter the discount percentage: "))

# Different logic: Directly calculate the final price
# If discount is 10%, the customer pays 90% of the price (100-10)/100 = 0.9
final_price = price * (1 - discount_percentage / 100)

print(f"The final price after a {discount_percentage}% discount is: ${final_price:.2f}")
```

```python
#Question 7: Check whether a person is eligible to vote.
#Input: Enter age
#Output: Eligible / Not Eligible

def check_voting_eligibility(age):
    """
    Checks if a person is eligible to vote based on age.

    Args:
        age (int): The age of the person.

    Returns:
        str: 'Eligible' if age is 18 or above, 'Not Eligible' otherwise.
    """
    if age >= 18:
        return "Eligible"
    else:
        return "Not Eligible"

# Get input from the user
person_age = int(input("Enter the age: "))

# Check eligibility and print the result
eligibility_status = check_voting_eligibility(person_age)
print(eligibility_status)
```

```python
#Question 8: Calculate electricity bill based on units consumed.
#Input: Enter units
#Output: Total bill amount

units = float(input("Enter the units consumed: "))

total_bill = 0.0

# Define new slab rates (example rates for 'new logic')
if units <= 100:
    total_bill = units * 3.00
elif units <= 300:
    # First 100 units at 3.00, remaining at 4.50
    total_bill = (100 * 3.00) + ((units - 100) * 4.50)
else:
    # First 100 units at 3.00, next 200 units at 4.50, remaining at 6.00
    total_bill = (100 * 3.00) + (200 * 4.50) + ((units - 300) * 6.00)

print(f"The total electricity bill for {units} units is: ${total_bill:.2f}")
```

```python
#Question 9: Print all even numbers between 1 to N.
#Input: Enter N
#Output: Even numbers list

N = int(input("Enter a number (N): "))

print(f"Even numbers between 1 and {N}:")
# Different logic: Using a list comprehension
even_numbers = [i for i in range(1, N + 1) if i % 2 == 0]

for num in even_numbers:
    print(num)
```
